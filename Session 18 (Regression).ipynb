```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv("D:\RegEX\DATASETS\Startups.csv")
```

```python
df.head()
```

```python
df.nunique()
```

```python
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
```

```python
X
```

```python
Y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(), [3])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
X
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X, Y)    # Higher the score better and error lower the better
```

```python
result = reg.predict(X)
```

```python
plt.figure(figsize = (10, 10))
plt.plot(Y, color = 'blue', label = 'Actual Values')
plt.plot(result, color = 'red', label = 'Predicted Values')
plt.legend()
```

```python
test_df = pd.DataFrame()
test_df['Actual_values'] = Y
test_df['Predicted_values'] = result
test_df
```

### Polynomial Regression
- Goal is to capture non linear relationship for improving flexibility of model

```python
df = pd.read_csv("D:\RegEX\DATASETS\Position_Salaries.csv")
```

```python
df.head(10)
```

```python
plt.figure()
plt.plot(df['Level'], df['Salary'])
```

```python
X = df[['Level']].values
Y = df[['Salary']].values
```

```python
X
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, Y)
```

```python
y_pred = reg.predict(X)
```

```python
plt.figure(figsize = (7, 7))
plt.plot(Y, color = 'blue', label = 'Actual Values')
plt.plot(y_pred, color = 'red', label = 'Predicted Values')
plt.legend()
```

```python
reg.score(X, Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 3)
X_poly = poly.fit_transform(X)
X_poly
```

```python
poly_reg = LinearRegression()
poly_reg.fit(X_poly, Y)
```

```python
y_pred2 = poly_reg.predict(X_poly)
```

```python
plt.figure(figsize = (7, 7))
plt.plot(Y, color = 'blue', label = 'Actual Values')
plt.plot(y_pred2, color = 'red', label = 'Predicted Values')
plt.legend()
```

```python
poly_reg.score(X_poly, Y)
```

```python
poly = PolynomialFeatures(degree = 5)
X_poly5 = poly.fit_transform(X)
X_poly5
```

```python
poly_reg5 = LinearRegression()
poly_reg5.fit(X_poly5, Y)
```

```python
y_pred5 = poly_reg5.predict(X_poly5)
```

```python
plt.figure(figsize = (7, 7))
plt.plot(Y, color = 'blue', label = 'Actual Values')
plt.plot(y_pred5, color = 'red', label = 'Predicted Values')
plt.legend()
```

```python
poly_reg5.score(X_poly5, Y)  # case of overfitting
```
