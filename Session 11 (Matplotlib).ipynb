# Types of Plots

## Line Plot

```python
# Line plot
plt.figure()
x = np.arange(10)
y = np.arange(10) * 2
plt.plot(x, y, 'r-*')
```

```python
plt.figure(figsize = (8, 8))
plt.plot(np.random.randn(30), 'mo-', label = 'FirstPlot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.legend()
```

```python
plt.figure()
x = np.arange(10)
y = x ** 2
z = x ** 3
plt.plot(x, y, 'mo-', label = "Squares")
plt.plot(x, z, 'b*-', label = 'Cubes')
plt.title("Squares and Cubes")
plt.legend()
```

## Scatter Plot

```python
# Scatter plot

plt.figure()
x = np.arange(10)
y = x ** 2
plt.scatter(x, y, s = 200, c = 'm')
```

```python
plt.xticks(np.arange(10))
```

```python
plt.yticks(np.arange(0, 100, 5))
```

## Bar Plots

```python
# Bar plots

plt.figure()
x = np.arange(10)
y = x * 2
plt.bar(x, y, width = 0.8)
plt.xticks(x)
plt.yticks(y)
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
z = x * 4
plt.bar(x,y, width = 0.4, color = 'm')
plt.bar(x,z, width = 0.4, color = 'c', bottom = x)
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
z = x * 4
plt.bar(x - 0.3, y, width = 0.4, color = 'm')
plt.bar(x + 0.2, z, width = 0.4, color = 'c')
```

## Histogram

```python
plt.figure()
plt.hist([1,2,3,4,5,4,2,5,6,7,8,9,5,6,7,0,4,3,5,6,7], width = 0.8)
plt.xticks(range(10))
plt.yticks(range(5))
```

```python
plt.figure()
x = np.random.normal(100, 10, 500)
plt.hist(x)
```

## Pie Chart

```python
plt.figure()
x = np.array([10, 20, 30, 80])
plt.pie(x)
```

```python
plt.figure()
x = [30, 20, 40, 30, 10]
food = ['Bananas', 'Apples', 'Grapes', 'Chickoo', 'Mango']
plt.pie(x, labels = food)
```
